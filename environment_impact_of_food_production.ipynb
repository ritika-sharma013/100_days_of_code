```python
import kagglehub
import pandas as pd
import os

# 1. Download the dataset
path = kagglehub.dataset_download("selfvivek/environment-impact-of-food-production")
print("Data downloaded to:", path)

# 2. List the files inside the downloaded folder
files = os.listdir(path)
print("Files in dataset:", files)

# 3. Load the primary CSV file (assuming it's the first CSV file)
csv_files = [f for f in files if f.endswith('.csv')]
if csv_files:
    df = pd.read_csv(os.path.join(path, csv_files[0]))
    print("\n Dataset is as follows:")
    display(df.iloc[:])
else:
    print("No CSV files found in the dataset folder.")
```
